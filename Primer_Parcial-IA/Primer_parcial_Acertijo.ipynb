```python
class Nodo:
    def __init__(self, estado, hijo=None):
        self.estado = estado
        self.hijo = None
        self.padre = None
        self.accion = None
        self.acciones = None
        self.costo = None
        self.set_hijo(hijo)

    def set_estado(self, estado):
        self.estado = estado

    def get_estado(self):
        return self.estado

    def set_hijo(self, hijo):
        self.hijo = hijo
        if self.hijo is not None:
            for s in self.hijo:
                s.padre = self

    def get_hijo(self):
        return self.hijo
    
    def set_padre(self, padre):
        self.padre = padre

    def get_padre(self):
        return self.padre
    
    def set_accion(self, accion):
        self.padre = accion

    def get_accion(self):
        return self.accion

    def set_acciones(self, acciones):
        self.acciones = acciones

    def get_acciones(self):
        return self.acciones

    def set_costo(self, costo):
        self.costo = costo

    def get_costo(self):
        return self.costo

    def equal(self, Nodo):
        if self.get_estado() == Nodo.get_estado():
            return True
        else:
            return False

    def en_lista(self, lista_nodos):
        enlistado = False
        for n in lista_nodos:
            if self.equal(n):
                enlistado = True
        return enlistado

    def __str__(self):
        return str(self.get_estado())
```

```python
#lista = ['_', '_', 'lechuga', 'persona', 'Lobo', 'oveja', '_', '_']
#lista2= ['Lobo', '_', 'lechuga', 'persona', '_', 'oveja', '_', '_']
# ['_', '_', 'lechuga', 'persona', 'Lobo', 'oveja', '_', '_']
#lado1: lechuga y persona, lado2: lobo y oveja
#lado1: lechuga y oveja, lado2 lobo y persona
#lado1:lechuga, oveja y Lobo, lado2: persona

def muertos(hijo):
    lado1 = [0,0,0,0]
    lado2 = [0,0,0,0]

#lado1: 
# idx 0: lechuga,
# idx 1: oveja,
# idx 2: lobo,
# idx 3: persona
    estados_muerto= [[1,1,0,0], #[lechuga, oveja, _ ,_ ] oveja de come a lechuga
                 [0,1,1,0], #[_, oveja, Lobo,_ ] lobo se come a oveja
                 [1,1,1,0]] #[lechuga, oveja, lobo ] oveja se come a lechuga y lobo a oveja
#['_', '_', '_', 'persona', 'Lobo', 'oveja', 'lechuga', '_']
    #index= obtiene la posicion de la lista 
    #v= el valor
    for index,v in enumerate(hijo):
        if index <4:
            if(v == "Lobo"): # si lobo esta en la posicion [2] se marca 1
                lado1[2] = 1
            elif(v == "lechuga"): # si lechuga esta en la posocion [0] se marca 1
                lado1[0] = 1
            elif(v == "persona"): # si Persona esta en la posicion [3] se marca 1
                lado1[3] = 1
            elif(v == "oveja"): # si oveja esta en la posicion [1] se marca 1
                lado1[1] = 1
        else :
            if(v == "Lobo"):
                lado2[2] = 1
            elif(v == "lechuga"):
                lado2[0] = 1
            elif(v == "persona"):
                lado2[3] = 1
            elif(v == "oveja"):
                lado2[1] = 1
    
    mueto=False;
    if(lado1 in estados_muerto): # si lado1 esta en estados muerto sera verdad
        mueto=True
    elif(lado2 in estados_muerto): # si lado2 esta en estdos muertos sera verdad
        mueto=True
    if mueto:
        return True 
    else:
        return False
```

```python
def busqueda_BPA_solucion(estado_inicial, solucion):
    resuelto = False
    nodos_visitados = []
    nodos_frontera = []

    nodo_raiz = Nodo(estado_inicial)
    nodos_frontera.append(nodo_raiz)
    while (not resuelto) and len(nodos_frontera) != 0:
        nodo_actual = nodos_frontera.pop()

        # extraer nodo y aÃ±adirlo a visitados
        nodos_visitados.append(nodo_actual)
        if nodo_actual.get_estado() == solucion:
            # solucion encontrada
            resuelto = True
            return nodo_actual
        else:
            # expandir nodos hijo
            estado_nodo = nodo_actual.get_estado()
            lista_hijos =[]       
            # operador izquierdo
                    #[lado1: _, lobo[1], oveja[2], persona[3]; lado2: [lechuga[0], _ ,_ , _ ]
            hijo = [estado_nodo[4], estado_nodo[1], estado_nodo[2], estado_nodo[3], estado_nodo[0], estado_nodo[5], estado_nodo[6], estado_nodo[7]]
            hijo_1 = Nodo(hijo)
            if(muertos(hijo)==False):
                lista_hijos.append(hijo_1)
            if not hijo_1.en_lista(nodos_visitados) and not hijo_1.en_lista(nodos_frontera):
                nodos_frontera.append(hijo_1)

            # operador central
            hijo = [estado_nodo[0], estado_nodo[5], estado_nodo[2], estado_nodo[3], estado_nodo[4], estado_nodo[1], estado_nodo[6], estado_nodo[7]]
            hijo_2 = Nodo(hijo)
            if(muertos(hijo)==False):
                lista_hijos.append(hijo_2)
            
            if not hijo_2.en_lista(nodos_visitados) and not hijo_2.en_lista(nodos_frontera):
                nodos_frontera.append(hijo_2)

            # operador derecho
            hijo = [estado_nodo[0], estado_nodo[1], estado_nodo[6], estado_nodo[3], estado_nodo[4], estado_nodo[5], estado_nodo[2], estado_nodo[7]]
            hijo_3 = Nodo(hijo)
            if(muertos(hijo)==False):
                
                lista_hijos.append(hijo_3)
            if not hijo_3.en_lista(nodos_visitados) and not hijo_3.en_lista(nodos_frontera):
                nodos_frontera.append(hijo_3)

            hijo = [estado_nodo[0], estado_nodo[1], estado_nodo[2], estado_nodo[7], estado_nodo[4], estado_nodo[5], estado_nodo[6], estado_nodo[3]]
            hijo_4 = Nodo(hijo)
            if(muertos(hijo)==False):
                
                lista_hijos.append(hijo_4)
            if not hijo_4.en_lista(nodos_visitados) and not hijo_4.en_lista(nodos_frontera):
                nodos_frontera.append(hijo_4)

            nodo_actual.set_hijo(lista_hijos)
```

```python
if __name__ == "__main__":
    # Lado1: 0 -3 lado2: 4-7
    estado_inicial = ["oveja", "Lobo", "lechuga", "persona", "_","_","_","_"]
    solucion = ["_","_","_","_","oveja", "Lobo", "lechuga", "persona"]
    nodo_solucion = busqueda_BPA_solucion(estado_inicial, solucion)
    # mostrar resultado
    resultado = []
    nodo_actual = nodo_solucion
    while nodo_actual.get_padre() is not None:
        resultado.append(nodo_actual.get_estado())
        nodo_actual = nodo_actual.get_padre()

    resultado.append(estado_inicial)
    resultado.reverse()
    print(resultado)
```
